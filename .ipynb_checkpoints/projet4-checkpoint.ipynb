### Convertir la colonne date en datetime, créer la colonne age et réorganiser le dataset

```python
# Convertir la colonne date en datetime
df['date'] = pd.to_datetime(df['date'])

# Créer la colonne age et la convertir avec la colonne categ en int
df['age'] = df['date'].dt.year - df['birth']
df['age'] = df['age'].astype(int)
df['categ'] = df['categ'].astype(int)

# Remplacer les lettres par des nombres dans la colonne sex et la convertir en int
#df['sex'].replace(['f','m'], [0, 1], inplace=True)
#df['sex'] = df['sex'].astype(int)

# Réorganiser le dataset en supprimer les colonnes inutiles pour l'analyse
df = df.reindex(columns=['client_id', 'session_id', 'sex','age','date','id_prod','categ','price'])
```

```python
# Création des colonnes year, monthn heure, day et nom_mois
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day

df['n_mois'] = df['date'].map(lambda d: d.month)
df['heure'] = df['date'].map(lambda d: d.hour)
```

```python
df.head()
```

```python
# Ecrire une fonction qui scinde la journée en Afternoon, Morning et Night

def partie_jour(value):
    if 14 <= value < 22:
        return 'Afternoon'
    elif 6 <= value < 14:
        return 'Morning'
    else:
        return 'Night'

df['partie_jour'] = df['heure'].map(partie_jour)
```

```python
# Ecrire une fonction qui scinde l'âge en tranches

def tranche_age(value):
    if 1 <= value < 20:
        return '1 - 19 ans'
    elif 19 <= value < 40:
        return '20 - 39 ans'
    elif 39 <= value < 60:
        return '40 - 59 ans'
    else:
        return '60 et plus'

df['tranche_age'] = df['age'].map(tranche_age)
```

### Identifier les ages aberrants pour chaque catégorie de produit puis les supprimer

```python
# Vérifier la présence ou non de valeurs aberrantes avec la fonction find_outliers
def find_outliers(v):
    Q1 = v.quantile(0.25)
    Q3 = v.quantile(0.75)
    IQ = Q3 - Q1
    LI = Q1 - (IQ*1.5)
    LS = Q3 + (IQ*1.5)
    i = list(v.index[(v < LI) | (v > LS)])
    val = list(v[i])
    return val
```

```python
# Prix aberrants en fonction de la catégorie
categ0 = find_outliers(df.loc[df['categ'] == 0, 'price'])
categ1 = find_outliers(df.loc[df['categ'] == 1, 'price'])
categ2 = find_outliers(df.loc[df['categ'] == 2, 'price'])
```

```python
# Âge extrême en fonction de la catégorie
categ0_age = find_outliers(df.loc[df['categ'] == 0, 'age'])
categ1_age = find_outliers(df.loc[df['categ'] == 1, 'age'])
categ2_age = find_outliers(df.loc[df['categ'] == 2, 'age'])
```

```python
#Répartition des tarifs par catégorie
data_boxplot = []
append_liste = []

for cat in df['categ'].unique():
    subset = df[df.categ == cat]
    data_boxplot.append(subset['age'])
    append_liste.append([int(cat), round(subset['age'].mean(),2),
                         round(subset['age'].median(),2),
                         round(subset['age'].std(ddof=0),2),
                         round(subset['age'].var(ddof=0),2)])
    
fig, ax1 = plt.subplots(figsize=(12, 8))
fig.subplots_adjust(left=0.075, right=0.95, top=0.9, bottom=0.25)

bp = ax1.boxplot(data_boxplot, notch=0, vert=1, whis=1.5)

ax1.yaxis.grid(True, linestyle='-', which='major', color='lightgrey',
               alpha=0.4)
ax1.set_axisbelow(True)
ax1.set_title('Distribution des ages des clients par catégorie', fontsize=22)
ax1.set_ylabel('Âge des clients')
ax1.set_xlim(0, len(data_boxplot) + 0.5)
top = 310
bottom = 0
ax1.set_ylim(bottom, top)


plot_legend = ['Catégorie 0', ' Catégorie 1', 'Catégorie 2']
ax1.set_xticklabels(plot_legend, fontsize=12)

plt.show()
```

## Analyse de deux variables quantatives (age et price)

```python
#Chiffre d'affaire par client
chiffre_daffaire_client = df.groupby('client_id').agg('sum').reset_index()
chiffre_daffaire_client = chiffre_daffaire_client[['client_id','price']]
chiffre_daffaire_client.rename(columns={'price':'chiffre_daffaire'}, inplace=True)
```

```python
#On fait la jointure df pour avoir l'âge du client dans notre dataset
chiffre_daffaire_client = pd.merge(chiffre_daffaire_client, df, on = ['client_id'], how = 'left')

# On choisit les colonnes utiles et drop les doublons
chiffre_daffaire_client = chiffre_daffaire_client[['client_id', 'chiffre_daffaire', 'sex', 'age']]
chiffre_daffaire_client = chiffre_daffaire_client.drop_duplicates(subset='client_id', keep='first')
```

```python
# courbe de Lorenz

ca = chiffre_daffaire_client['chiffre_daffaire'].values

lorenz = np.cumsum(np.sort(ca)) / ca.sum()
lorenz = np.append([0], lorenz)

plt.figure(figsize=(12,12))
plt.plot(np.linspace(0, 1, len(lorenz)), lorenz, drawstyle = 'steps-post')

plt.gca().set_aspect('equal', adjustable = 'box')
labels =['Courbe de concentration', 'Bissectrice : égalité parfaite']

x = [0, 1]
y = [0, 1]

plt.plot(x, y, 'red')
plt.ylabel("Chiffre d'affaires", size=15)
plt.xlabel("Clients", size=15)


plt.ylim(0, 1)
plt.xlim(0, 1)

plt.title('Proportion du chiffre par rapport aux clients', fontsize=(18),fontweight='bold')
plt.legend(labels)

plt.plot(0.8, 0.52, 'o')
plt.text(0.81, 0.49 , "80% des clients génèrent \n52% du chiffre d'affaire", 
                                                                     fontsize = '18')

plt.grid()                                                                                    
plt.show()

aire_courbe = lorenz[:-1].sum() / len(lorenz)

S = 0.5 - aire_courbe
gini = 2 * S
print('L\'indice de Gini est:', round(gini,2))

```

```python
#Chiffre d'affaire par produit
chiffre_daffaire_prod = df.groupby('id_prod').agg('sum').reset_index()
chiffre_daffaire_prod = chiffre_daffaire_prod[['id_prod','price']]
chiffre_daffaire_prod.rename(columns={'price':'chiffre_daffaire'}, inplace=True)
```

```python
# courbe de Lorenz

ca = chiffre_daffaire_prod['chiffre_daffaire'].values

lorenz = np.cumsum(np.sort(ca)) / ca.sum()
lorenz = np.append([0], lorenz)

plt.figure(figsize=(12,12))
plt.plot(np.linspace(0, 1, len(lorenz)), lorenz, drawstyle = 'steps-post')

plt.gca().set_aspect('equal', adjustable = 'box')
labels =['Courbe de concentration', 'Bissectrice : égalité parfaite']

x = [0, 1]
y = [0, 1]

plt.plot(x, y, 'red')
plt.ylabel("Chiffre d'affaires", size=15)
plt.xlabel("Produits", size=15)


plt.ylim(0, 1)
plt.xlim(0, 1)

plt.title('Proportion du chiffre par rapport aux produits', fontsize=(18),fontweight='bold')
plt.legend(labels)

plt.plot(0.8, 0.22, 'o')
plt.text(0.82, 0.2 , "80% des produits génèrent \n22% du chiffre d'affaire", 
                                                                     fontsize = '18')

plt.grid()                                                                                    
plt.show()

aire_courbe = lorenz[:-1].sum() / len(lorenz)

S = 0.5 - aire_courbe
gini = 2 * S
print('L\'indice de Gini est:', round(gini,2))

```

```python
fig, ax = plt.subplots(figsize=(12,8))
taille_classe = 10 # taille des classes pour la discrétisation

groupes = [] # va recevoir les données agrégées à afficher

# on calcule des tranches allant de 0 au solde maximum par paliers de taille taille_classe
tranches = np.arange(0, max(df["age"]), taille_classe)
#tranches += taille_classe/2 # on décale les tranches d'une demi taille de classe
indices = np.digitize(df["age"], tranches) # associe chaque age à son numéro de classe

for ind, tr in enumerate(tranches): # pour chaque tranche, ind reçoit le numéro de tranche et tr la tranche en question
    age = df.loc[indices==ind,"price"] # sélection des individus de la tranche ind
    if len(age) > 0:
        g = {
            'valeurs': age,
            'centre_classe': tr-(taille_classe/2),
            'taille': len(age),
            'quartiles': [np.percentile(age,p) for p in [25,50,75]]
        }
        groupes.append(g)

# Propriétés graphiques (pas très importantes)    
medianprops = {'color':"black"}
meanprops = {'marker':'o', 'markeredgecolor':'black',
            'markerfacecolor':'firebrick'}        

# affichage des boxplots
plt.boxplot([g["valeurs"] for g in groupes],
            positions= [g["centre_classe"] for g in groupes], # abscisses des boxplots
            showfliers= False, # on ne prend pas en compte les outliers
            widths= taille_classe*0.7, # largeur graphique des boxplots
            medianprops=medianprops, 
            vert=True, patch_artist=True, showmeans=True, meanprops=meanprops
)
plt.xlabel('Age')
plt.ylabel('Montant du panier')
plt.title('Répartition des commandes par groupe d\'âge', 
  loc='center', 
  fontsize=25)

# affichage des effectifs de chaque classe
for g in groupes:
    plt.text(g["centre_classe"],0,"(n={})".format(g["taille"]),horizontalalignment='center',verticalalignment='top')     
plt.show()
```

```python
X = "tranche_age" # qualitative
Y = "mnt_panier" # quantitative

def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
    
np.round(eta_squared(mnt_panier[X],mnt_panier[Y]),2)
```

```python
fig, ax = plt.subplots(figsize=(12,8))
X = "categ" # qualitative
Y = "age" # quantitative

sous_echantillon = df.copy() # On ne garde que les dépenses
modalites = sous_echantillon[X].unique()
groupes = []
for m in modalites:
    groupes.append(sous_echantillon[sous_echantillon[X]==m][Y])

# Propriétés graphiques (pas très importantes)    
medianprops = {'color':"black"}
meanprops = {'marker':'o', 'markeredgecolor':'black',
            'markerfacecolor':'firebrick'}
plt.xlabel('Catégories', size=15)
plt.ylabel('Âge des clients', size=15)
plt.title("Pas de dépendance entre l'âge du client et la catégorie", size=22)   
plt.boxplot(groupes, labels=modalites, showfliers=False, medianprops=medianprops, 
            vert=True, patch_artist=True, showmeans=True, meanprops=meanprops)
plt.show()

def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])
    return SCE/SCT
    
print('Il n\'y a une certaine corrélation entre la catégorie et l\'âge du client: ',
      np.round(eta_squared(sous_echantillon[X],sous_echantillon[Y]),2))
```
